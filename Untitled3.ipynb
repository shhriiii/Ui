```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, mean_squared_error
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import roc_curve, roc_auc_score, RocCurveDisplay
from joblib import dump, load
import gradio as gr
```

```python
df = pd.read_csv("data.csv")
```

```python
null_var = df.isnull().sum()/df.shape[0]*100
drop_columns = null_var[null_var>20].keys()
df2_drop_clm = df.drop(columns=drop_columns)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
label_enco = LabelEncoder()
categorical_columns = df2_drop_clm.select_dtypes(include=['object', 'category']).columns
for col in categorical_columns:
    df2_drop_clm[col + '_encoded'] = label_enco.fit_transform(df[col])
```

```python
df2_drop_clm = df2_drop_clm.drop(columns=['Ball Rebowled', 'Extra Type_encoded', 'Date_encoded', 'Unnamed: 0'])
df2 = df2_drop_clm.drop(columns='Chased Successfully')
df2['New_Column'] = df2.apply(lambda row: 1 if row['Winner_encoded'] == row['Bat First_encoded'] else 0, axis=1)
df2.drop(columns=['Batter Runs', 'Extra Runs', 'Runs From Ball'], inplace=True)
```

```python
X = df2.drop(columns=['Winner_encoded', 'Match ID', 'Innings', 'Non Striker_encoded', 'Date', 'Venue', 'Bat First', 'Bat Second', 'Batter', 'Non Striker',
       'Bowler', 'Extra Type', 'Winner', 'New_Column', 'Venue_encoded'], axis=1)
y = df2['New_Column']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=51)
```

```python
model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_lr.fit(X_train, y_train)
```

```python
y_predicted = model_lr.predict(X_test)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_predicted)}')
mse = mean_squared_error(y_test, y_predicted)
print(f'Mean-Squared Error: {mse}')
```

```python
conf_matrix = confusion_matrix(y_test, y_predicted)
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], 
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.show()
```

```python
print("Classification Report:")
print(classification_report(y_test, y_predicted))
```

```python
precision = precision_score(y_test, y_predicted)
recall = recall_score(y_test, y_predicted)
f1 = f1_score(y_test, y_predicted)
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

```python
y_proba = model_lr.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_proba)
auc = roc_auc_score(y_test, y_proba)
RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=auc).plot()
plt.show()
```

```python
dump(model_lr, 'model_lr.joblib')

# Load the model (for demonstration, in practice you wouldn't need to load right after saving)
model = load('model_lr.joblib')
```

```python
def predict_winner(over, ball, wicket, innings_runs, innings_wickets, target_score, balls_remaining,
                   total_batter_runs, total_non_striker_runs, batter_balls_faced, non_striker_balls_faced,
                   bowler_runs_conceded, valid_ball, bat_first, bat_second, batter, bowler):
    
    input_data = pd.DataFrame({
        'Over': [over],
        'Ball': [ball],
        'Wicket': [wicket],
        'Innings Runs': [innings_runs],
        'Innings Wickets': [innings_wickets],
        'Target Score': [target_score],
        'Balls Remaining': [balls_remaining],
        'Total Batter Runs': [total_batter_runs],
        'Total Non Striker Runs': [total_non_striker_runs],
        'Batter Balls Faced': [batter_balls_faced],
        'Non Striker Balls Faced': [non_striker_balls_faced],
        'Bowler Runs Conceded': [bowler_runs_conceded],
        'Valid Ball': [valid_ball],
        'Bat First_encoded': [label_enco.transform([bat_first])[0]],
        'Bat Second_encoded': [label_enco.transform([bat_second])[0]],
        'Batter_encoded': [label_enco.transform([batter])[0]],
        'Bowler_encoded': [label_enco.transform([bowler])[0]]
    })

    prediction = model.predict(input_data)
    probability = model.predict_proba(input_data)[0]
    
    if prediction[0] == 1:
        result = f"Team batting first is predicted to win with probability {probability[1]:.2f}"
    else:
        result = f"Team batting second is predicted to win with probability {probability[0]:.2f}"
    
    return result
```

```python
iface = gr.Interface(
    fn=predict_winner,
    inputs=[
        gr.Number(label="Over"),
        gr.Number(label="Ball"),
        gr.Number(label="Wicket"),
        gr.Number(label="Innings Runs"),
        gr.Number(label="Innings Wickets"),
        gr.Number(label="Target Score"),
        gr.Number(label="Balls Remaining"),
        gr.Number(label="Total Batter Runs"),
        gr.Number(label="Total Non Striker Runs"),
        gr.Number(label="Batter Balls Faced"),
        gr.Number(label="Non Striker Balls Faced"),
        gr.Number(label="Bowler Runs Conceded"),
        gr.Number(label="Valid Ball"),
        gr.Dropdown(choices=df['Bat First'].unique().tolist(), label="Bat First"),
        gr.Dropdown(choices=df['Bat Second'].unique().tolist(), label="Bat Second"),
        gr.Dropdown(choices=df['Batter'].unique().tolist(), label="Batter"),
        gr.Dropdown(choices=df['Bowler'].unique().tolist(), label="Bowler")
    ],
    outputs="text",
    title="Cricket Match Winner Prediction",
    description="Enter match details to predict which team is likely to win"
)

iface.launch()
```
